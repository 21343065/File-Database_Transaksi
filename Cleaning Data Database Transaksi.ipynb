```python
import mysql.connector
import matplotlib.pyplot as plt

# Membuat koneksi ke MySQL
mydb = mysql.connector.connect(
    host="",
    user="root",
    password="",
    database="transaksi"
)

# Membuat objek cursor
cursor = mydb.cursor()

# Query untuk mengambil data transaksi
query = " SELECT customer_grid_view.domisili AS kota, SUM(transaksi_grid_view1.Harga_Asli) AS total_transaksi FROM transaksi_grid_view1 JOIN customer_grid_view ON transaksi_grid_view1.customer = customer_grid_view.Name GROUP BY customer_grid_view.domisili;"

# Menjalankan query
cursor.execute(query)

# Mengambil hasil query
results = cursor.fetchall()

# Menyiapkan data untuk visualisasi
customer = [row[0] for row in results]
total_transaksi = [row[1] for row in results]

# Membuat grafik bar
plt.bar(customer, total_transaksi)
plt.xlabel('Customer')
plt.ylabel('Total Transaksi')
plt.title('Grafik Transaksi')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()

# Menampilkan grafik
plt.show()

# Menutup koneksi dan cursor
cursor.close()
mydb.close()
```

```python
import mysql.connector
import matplotlib.pyplot as plt

# Membuat koneksi ke MySQL
mydb = mysql.connector.connect(
    host="",
    user="root",
    password="",
    database="transaksi"
)

# Membuat objek cursor
cursor = mydb.cursor()

# Query untuk mengambil data campaign
query = "SELECT Name, SUM(Budget) AS total_budget FROM Campaign_grid_view GROUP BY Name;"

# Menjalankan query
cursor.execute(query)

# Mengambil hasil query
results = cursor.fetchall()

# Menyiapkan data untuk visualisasi
Name = [row[0] for row in results]
total_budget = [row[1] for row in results]

# Membuat grafik bar
plt.bar(Name, total_budget)
plt.xlabel('Tipe Campaign')
plt.ylabel('Total Budget')
plt.title('Grafik Total Budget per Tipe Campaign')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Menampilkan grafik
plt.show()

# Menutup koneksi dan cursor
cursor.close()
mydb.close()
```
